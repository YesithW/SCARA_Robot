```python
import numpy as np
import matplotlib.pyplot as plt

# Define SCARA parameters
L1 = 10.0  # Length of first link
L2 = 10.0  # Length of second link
theta1_min, theta1_max = np.radians([0, 45])  # Joint 1 limits
theta2_min, theta2_max = np.radians([-45, 45])  # Joint 2 limits

# Define resolution
theta1_values = np.linspace(theta1_min, theta1_max, 100)
theta2_values = np.linspace(theta2_min, theta2_max, 100)

reachable_x = []
reachable_y = []

# Compute reachable positions
for theta1 in theta1_values:
    for theta2 in theta2_values:
        x = L1 * np.cos(theta1) + L2 * np.cos(theta1 + theta2)
        y = L1 * np.sin(theta1) + L2 * np.sin(theta1 + theta2)
        reachable_x.append(x)
        reachable_y.append(y)

# Plot the workspace
plt.figure(figsize=(6,6))
plt.scatter(reachable_x, reachable_y, s=1, color='blue')
plt.xlabel("X Position")
plt.ylabel("Y Position")
plt.title("SCARA Robot Workspace")
plt.axis("equal")
plt.grid(True)
plt.show()
```
